```python
import matplotlib.pyplot as plt
import numpy as np

from skimage.io import imread
from skimage.color import rgb2ycbcr
```

```python
# Importar uma foto de um gatinho
url = "https://placekitten.com/200/200"
rgb_img = imread(url)

height, width, layers = rgb_img.shape

# Matriz de zeros, para zerar os canais
z = np.zeros((height, width), dtype="uint8")

# Criar uma cpia zerando canais G e B
r_img = rgb_img.copy()
r_img[:, :, 1] = z
r_img[:, :, 2] = z

# Criar uma cpia zerando canais R e B
g_img = rgb_img.copy()
g_img[:, :, 0] = z
g_img[:, :, 2] = z

# Criar uma cpia zerando canais B e R
b_img = rgb_img.copy()
b_img[:, :, 0] = z
b_img[:, :, 1] = z

# Plotar a imagem original
plt.figure()
plt.imshow(rgb_img)

# Plotar o canal R
plt.figure()
plt.imshow(r_img)

# Plotar o canal G
plt.figure()
plt.imshow(g_img)

# Plotar o canal B
plt.figure()
plt.imshow(b_img)
```

```python
# Converter imagem para YCbCr
ycc_img = rgb2ycbcr(rgb_img)

# Serarar os canais
img_y  = ycc_img[:, :, 0]
img_cb = ycc_img[:, :, 1]
img_cr = ycc_img[:, :, 2]

# Plotar os canais separados
plt.figure()
plt.imshow(img_y, cmap=plt.cm.gray)

plt.figure()
plt.imshow(img_cb, cmap=plt.cm.gray)

plt.figure()
plt.imshow(img_cr, cmap=plt.cm.gray)
```
